```python
import math
import sys
class Arithmatic:
    def _init_(self):
        self.x=0
        self.y=0
        self.res=none

    def add(self):
        self.x=int(input("Enter x value: "))
        self.y=int(input("Enter y value: "))
        self.res=self.x+self.y
        print("Addition is ",self.res)

    def sub(self):
        self.x=int(input("Enter x value: "))
        self.y=int(input("Enter y value: "))
        self.res=self.x-self.y
        print("Substraction is ",self.res)

    def mult(self):
        self.x=int(input("Enter x value: "))
        self.y=int(input("Enter y value: "))
        self.res=self.x*self.y
        print("Multiplication is ",self.res)

    def div(self):
        self.x=int(input("Enter x value: "))
        self.y=int(input("Enter y value: "))
        self.res=self.x/self.y
        print("Division is ",self.res)

    def area(self):
        self.x=int(input("Enter radius of circle: "))
        self.res=3.14*self.x*self.x
        print("Area of circle ",self.res)

    def circum(self):
        self.x=int(input("Enter radius: "))
        self.res=2*3.14*self.x
        print("Circumference is ",self.res)

    def rect(self):
        self.x=int(input("Enter length: "))
        self.y=int(input("Enter breadth: "))
        self.res=self.x*self.y
        print("Area of rectangle: ",self.res)


    def vol(self):
        self.x=int(input("Enter the radius: "))
        self.res=4/3*3.14*self.x*self.x*self.x
        print("Volume of sphere: ",self.res)


    def surf(self):
        self.x=int(input("Enter radius: "))
        self.res=4*3.14*self.x*self.x
        print("Surface area of sphere ",self.res)


    def squa(self):
        self.x=int(input("Enter side value: "))
        self.res=self.x*self.x
        print("Area of square ",self.res)

    def right(self):
        self.x=int(input("Enter Base: "))
        self.y=int(input("Enter Height: "))
        self.res=1/2*self.x*self.y
        print("Area of right angle triangle ",self.res)



    def equi(self):
        x=int(input("Enter side value: "))
        self.res= math.sqrt(3)/4 * self.x*self.x
        print("Area of equilateral triangle ",self.res)


    def peri(self):
        self.x=int(input("Enter Width: "))
        self.y=int(input("Enter Height: "))
        self.res=2*(self.x+self.y)
        print("Perimeter of Rectangle ",self.res)


    def tri(self):
        self.x=int(input("Enter Base: "))
        self.y=int(input("Enter Height: "))
        self.res=1/2*self.x*self.y
        print("Area of triangle ",self.res)


    def square(self):
        self.x=int(input("Enter Value: "))
        self.res=self.x*self.x
        print("Square: ",self.res)

    def cube(self):
        self.x=int(input("Enter Value: "))
        self.res=self.x*self.x * self.x
        print("Cube ",self.res)



obj=Arithmatic()
while True:
   print("\n\n1. Addition\t2. Substraction\t3. Multiplication\t4. Division")
   print("5. Area of Circle\t6. Circumference of Circle\t7. Area of Rectangle\t8. Volume of Sphere")
   print("9. Surface area of Sphere\t10. Area of Sphere\t11. Area of Right Angle Triangle\t12. Area of Equilatera Triangle")
   print("13. Perimeter of Rectangle\t14. Area of Triangle\t15. Squre of the no.\t16. Cube of the no.")
   print("\t0. Exit")
   ch=int(input("Select any choice..."))
   if ch==1:
    obj.add()
   elif ch==2:
    obj.sub()
   elif ch==3:
    obj.mult()
   elif ch==4:
    obj.div()
   elif ch==5:
    obj.area()
   elif ch==6:
    obj.circum()
   elif ch==7:
    obj.rect()
   elif ch==8:
    obj.vol()
   elif ch==9:
    obj.surf()
   elif ch==10:
    obj.squa()
   elif ch==11:
    obj.right()
   elif ch==12:
    obj.equi()
   elif ch==13:
    obj.peri()
   elif ch==14:
    obj.tri()
   elif ch==15:
    obj.square()
   elif ch==16:
    obj.cube()
   elif ch==0:
    sys.exit()
   else :
    print("Invalid Option")


obj.add()
obj.sub()
obj.mult()
obj.div()
obj.area()
obj.circum()
obj.rect()
obj.vol()
obj.surf()
obj.squa()
obj.right()
obj.equi()
obj.peri()
obj.tri()
obj.square()
obj.cube()

```
